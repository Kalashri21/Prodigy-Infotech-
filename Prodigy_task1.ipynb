```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
# URL of the dataset
url ='https://raw.githubusercontent.com/datasets/population/master/data/population.csv'

# Load the dataset
data = pd.read_csv(url)
```

```python
# inspect the first few rows of the dataset
print(data.head())

# Inspect the column names
print(data.columns)

```

```python
# Inspect the column names
print(data.columns)
```

```python
# Select relevent Columns
population_data = data[['Year','Value']]
population_data
```

```python
# create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(population_data['Year'], population_data['Value'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth')
plt.show
```

```python
# Create a histogram of the population values
plt.figure(figsize=(10, 6))
plt.hist(population_data['Value'], bins=10)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution')
plt.show()
```
